```python
import numpy as np
from PIL import Image
from matplotlib import pyplot as plt
```

```python
with Image.open('input_images/img6.jpg') as imp_img1:
    gray_image1 = imp_img1.convert("L")
    gray_image1_array = np.array(gray_image1)
    with Image.open('input_images/img7.jpg') as imp_img2:
        gray_image2 = imp_img2.convert("L")
        gray_image2_array = np.array(gray_image2)
        max_width= max(len(gray_image1_array[0]),len(gray_image2_array[0]))
        max_height = max(len(gray_image1_array),len(gray_image2_array))
        new_image = np.zeros((max_height,max_width),dtype=np.uint8)
        print(len(new_image),len(new_image[0]))
        for i in range(max_height):
            for j in range(max_width):
                try:
                    new_image[i][j] = gray_image1_array[i][j]
                except:
                    pass
                try:
                    new_image[i][j] = new_image[i][j] | gray_image2_array[i][j]
                except:
                    pass
    output_image = Image.fromarray(new_image)
    
    plt.subplot(2,2,1)
    plt.imshow(gray_image1)
    plt.title("Image 1")
    plt.axis('off')

    plt.subplot(2,2,2)
    plt.imshow(gray_image2)
    plt.title("Image 2")
    plt.axis('off')

    plt.subplot(2,2,3)
    plt.imshow(output_image)
    plt.title("Or Operation")
    plt.axis('off')

    plt.tight_layout()

    plt.show()
```
