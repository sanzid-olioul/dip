```python
import numpy as np
from PIL import Image
from matplotlib import pyplot as plt
```

```python
with Image.open('input_images/img6.jpg') as imp_img:
    gray_image = imp_img.convert("L")
    gray_image_array = np.array(gray_image)
    ratio = len(gray_image_array)/len(gray_image_array[0])

    for i in range(len(gray_image_array)):
        for j in range(len(gray_image_array[0])):
            if (i> ratio*j-25 and i< ratio*j+25) or ((i > len(gray_image_array) -25-ratio*j) and (i < len(gray_image_array) +25-ratio*j)):
                gray_image_array[i][j]=0
    diagonal_image = Image.fromarray(gray_image_array)
    plt.figure(figsize=(15,15))
    plt.subplot(1,2,1)
    plt.imshow(gray_image)
    plt.title("Original Image")
    plt.axis('off')

    plt.subplot(1,2,2)
    plt.imshow(diagonal_image)
    plt.title("Diagonal image")
    plt.axis('off')

    plt.tight_layout()

    plt.show()
```
