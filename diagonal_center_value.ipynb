```python
import numpy as np
import pandas as pd
from PIL import Image
from matplotlib import pyplot as plt
```

```python
with Image.open('input_images/img2.jpg') as input_img:
    gray_image = input_img.convert('L')
    gray_image_array = np.array(gray_image,dtype=np.uint8)
    output_image_arr = np.zeros_like(gray_image_array)
    for i in range(len(gray_image_array)):
        for j in range(len(gray_image_array[0])):
            up = gray_image_array[i-1][j] if i > 0 else 0
            down = gray_image_array[i+1][j] if i < len(gray_image_array) - 1 else 0
            left = gray_image_array[i][j-1] if j > 0 else 0
            right = gray_image_array[i][j+1] if j < len(gray_image_array[0]) - 1 else 0
            
            output_image_arr[i][j] += min(gray_image_array[i][j] + up, 255)
            output_image_arr[i][j] += min(gray_image_array[i][j] + down, 255)
            output_image_arr[i][j] += min(gray_image_array[i][j] + left, 255)
            output_image_arr[i][j] += min(gray_image_array[i][j] + right, 255)

    output_image = Image.fromarray(output_image_arr)
    
    plt.subplot(2,2,1)
    plt.imshow(gray_image)
    plt.title("Image 1")
    plt.axis('off')

    plt.subplot(2,2,2)
    plt.imshow(output_image)
    plt.title("Image 2")
    plt.axis('off')
```
