```python
import numpy as np
from PIL import Image
from matplotlib import pyplot as plt
```

```python
def image_substraction(gray_image1,gray_image2):
    gray_image2_array = np.array(gray_image2)
    gray_image1_array = np.array(gray_image1)
    max_width= max(len(gray_image1_array[0]),len(gray_image2_array[0]))
    max_height = max(len(gray_image1_array),len(gray_image2_array))
    new_image = np.zeros((max_height,max_width),dtype=np.uint8)
    print(len(new_image),len(new_image[0]))
    for i in range(max_width):
        for j in range(max_height):
            try:
                new_image[i][j] = gray_image1_array[i][j]
            except:
                pass
            try:
                if new_image[i][j] >gray_image2_array[i][j]:
                    new_image[i][j] -= gray_image2_array[i][j]
                else:
                    new_image[i][j] = 0
            except:
                pass
    output_image = Image.fromarray(new_image)
    return output_image
```

```python
def image_addition(gray_image1,gray_image2):
    gray_image2_array = np.array(gray_image2)
    gray_image1_array = np.array(gray_image1)
    max_width= max(len(gray_image1_array[0]),len(gray_image2_array[0]))
    max_height = max(len(gray_image1_array),len(gray_image2_array))
    new_image = np.zeros((max_height,max_width),dtype=np.uint8)
    print(len(new_image),len(new_image[0]))
    for i in range(max_height):
        for j in range(max_width):
            try:
                new_image[i][j] = gray_image1_array[i][j]
            except:
                pass
            try:
                if new_image[i][j] + gray_image2_array[i][j] < 255:
                    new_image[i][j] += gray_image2_array[i][j]
                
            except:
                pass
    output_image = Image.fromarray(new_image)
    return output_image
```

```python
def image_and(gray_image1,gray_image2):
    gray_image2_array = np.array(gray_image2)
    gray_image1_array = np.array(gray_image1)
    max_width= max(len(gray_image1_array[0]),len(gray_image2_array[0]))
    max_height = max(len(gray_image1_array),len(gray_image2_array))
    new_image = np.zeros((max_height,max_width),dtype=np.uint8)
    print(len(new_image),len(new_image[0]))
    for i in range(max_height):
        for j in range(max_width):
            try:
                new_image[i][j] = gray_image1_array[i][j]
            except:
                pass
            try:
                new_image[i][j] = new_image[i][j] & gray_image2_array[i][j]
            except:
                pass
    output_image = Image.fromarray(new_image)
    return output_image
```

```python
def image_or(gray_image1,gray_image2):
    gray_image2_array = np.array(gray_image2)
    gray_image1_array = np.array(gray_image1)
    max_width= max(len(gray_image1_array[0]),len(gray_image2_array[0]))
    max_height = max(len(gray_image1_array),len(gray_image2_array))
    new_image = np.zeros((max_height,max_width),dtype=np.uint8)
    print(len(new_image),len(new_image[0]))
    for i in range(max_height):
        for j in range(max_width):
            try:
                new_image[i][j] = gray_image1_array[i][j]
            except:
                pass
            try:
                new_image[i][j] = new_image[i][j] | gray_image2_array[i][j]
            except:
                pass
    output_image = Image.fromarray(new_image)
    return output_image
    
```

```python
with Image.open('input_images/img6.jpg') as imp_img1:
    gray_image1 = imp_img1.convert("L")
    with Image.open('input_images/img7.jpg') as imp_img2:
        gray_image2 = imp_img2.convert("L")
        addition_img = image_addition(gray_image1,gray_image2)
        substraction_img = image_substraction(gray_image1,gray_image2)
        and_img = image_and(gray_image1,gray_image2)
        or_img = image_or(gray_image1,gray_image2)

        plt.subplot(3,2,1)
        plt.imshow(gray_image1,cmap='gray')
        plt.title("Image 1")
        plt.axis('off')

        plt.subplot(3,2,2)
        plt.imshow(gray_image2,cmap='gray')
        plt.title("Image 2")
        plt.axis('off')

        plt.subplot(3,2,3)
        plt.imshow(addition_img,cmap='gray')
        plt.title("Addition")
        plt.axis('off')

        plt.subplot(3,2,4)
        plt.imshow(substraction_img,cmap='gray')
        plt.title("Substraction")
        plt.axis('off')

        plt.subplot(3,2,5)
        plt.imshow(and_img,cmap='gray')
        plt.title("And")
        plt.axis('off')

        plt.subplot(3,2,6)
        plt.imshow(or_img,cmap='gray')
        plt.title("Or")
        plt.axis('off')

        plt.tight_layout()

        plt.show()
```
